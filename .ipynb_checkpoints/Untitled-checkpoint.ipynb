```python
#matrice de regression 
import numpy as np
global p 
def prep_mat(X,h):
    """
    Arguments:
    X array
    h horizon positive integer 
    Returns 
    B list of lists 
    """
    p=len(X)-h
    M=np.zeros((p,h))
    for i in range(len(X)-h):
        l=X[len(X)-1-h-i:len(X)-i-1]
        M[i:]=l[::-1]
        B=np.transpose(M)
        C=B[:,0:p]

    return C




#matrice de la métrique

import math 
def metr_retard(a,b,p):
    M=np.eye(p)
    for j in range(p) :
          M[j,j]  = math.exp(a + j*b)
    return M

def metr(h):
    W=np.eye(h)
    for j in range(h) :
          W[j,j]  = 1/h
    return W
# regression ridge
def ridge(X,Y,l,h,a,b):
    
    global p
    """"
    Arguments : 
    (X,Y) :une serie temporelle
    l     :facteurs de pénalisation 
    h     : horizon 
    (a,b) :des reels positives pour construir la matrice metrique 
    """
    p=len(X)-h
    M=metr_retard(a,b,p)#métrique
    x=prep_mat(X,h)#la matrice X(t-1,t-h)
    W=metr(h)
    # reglage de la dimension de Y
    T=Y[len(X)-h:len(X)]
    y=T[::-1]#les h derniers coeficients de y
    #calcul de l'estimateur de ridge (X'MX+lM)^(-1)X'My
    J1=np.transpose(x).dot(W) 
    J2=J1.dot(x)
    B=np.linalg.inv(J2+l*M)
    b1=B.dot(np.transpose(x))
    b2=b1.dot(W)
    b3=b1.dot(y)
    return b3
```

```python
import pandas as pd
from pandas import DataFrame
%pwd
serieX=pd.read_csv('serie X.csv',sep='	') 
serieX
```

```python
Y=X
b=ridge(X,Y,1,9,1,1)
b
```
